### **Análise Exploratória**

```python
import pandas as pd
from tabulate import tabulate
import seaborn as sns
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
df = pd.read_csv('dados/base-youtube.csv', encoding='latin')
df
```

```python
df.describe().T
```

```python
df['Country'].value_counts()
```

```python
df['channel_type'].value_counts()
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=['Country'])
df = df.dropna(subset=['category'])
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df_10_canais_inscritos = df[['Youtuber', 'subscribers', 'video views', 'category', 'Country']].sort_values(by='subscribers', ascending=False).head(10)
df_10_canais_inscritos
```

```python
plt.figure(figsize=(12, 6))
plt.bar(df_10_canais_inscritos['Youtuber'], df_10_canais_inscritos['subscribers'], color='g')
for i, valor in enumerate(df_10_canais_inscritos['subscribers']):
    plt.text(i, valor + 0.05 * valor, f'{valor:,.0f}', ha='center', va='bottom', fontsize=10)

plt.xlabel('Canal Youtube')
plt.ylabel('Inscritos')
plt.title('Os 10 maiores Canais do Youtube')
plt.xticks(rotation=75)
plt.tight_layout()
plt.show()
```

```python
df_10_canais_visualizacoes = df[['Youtuber', 'subscribers', 'video views', 'category', 'Country']].sort_values(by='video views', ascending=False).head(10)
df_10_canais_visualizacoes
```

```python
df_brasil = df[df['Country'] == 'Brazil']
df_top5_brasil = df_brasil[['Youtuber', 'video views', 'subscribers', 'category', ]].sort_values(by='video views', ascending=False).head(10)
print("Top 10 Canais Mais Assistidos do Brasil:")
print(df_top5_brasil)
```

```python
df_canais_musica = df[df['category'] == 'Music']
df_rank_musica = df_canais_musica[['Youtuber', 'subscribers', 'video views', 'Country',]].sort_values(by='subscribers', ascending=False).head(10)
df_rank_musica.reset_index(drop=True, inplace=True)
df_rank_musica
```

```python
top_20_music_channels = (
    df[df['category'] == 'Music']
    .sort_values(by='video views', ascending=False)  # Ordenar por visualizações em ordem decrescente
    .head(20)  # Selecionar os 20 primeiros
    [['Youtuber', 'subscribers', 'video views', 'Country']]  # Colunas relevantes
)

# Exibir o resultado
print(top_20_music_channels)
```

```python
cpm = 1.75  
df_canais_top10 = df[['Youtuber', 'video views', 'category', 'Country']].sort_values(by='video views', ascending=False)
def calcular_ganho_est(visualizacoes, cpm):
    return (visualizacoes / 1000) * cpm

df_canais_top10['Ganho Estimado (USD)'] = df_canais_top10['video views'].apply(lambda x: calcular_ganho_est(x, cpm))

print(df_canais_top10)
```

```python
cpm_brasil = 1.75 

df_canais_brasil = df[df['Country'] == 'Brazil']
df_canais_brasil_top10 = df_canais_brasil[['Youtuber', 'video views']].sort_values(by='video views', ascending=False)

def calcular_ganho_est(visualizacoes, cpm):
    return (visualizacoes / 1000) * cpm

df_canais_brasil_top10['Ganho Estimado (USD)'] = df_canais_brasil_top10['video views'].apply(lambda x: calcular_ganho_est(x, cpm_brasil))

print(df_canais_brasil_top10)
```

```python
df_corr = df[['Country', 'video views', 'Gross tertiary education enrollment (%)']].dropna()

plt.figure(figsize=(12, 6))
sns.scatterplot(x='Country', y='video views', hue='Gross tertiary education enrollment (%)', data=df_corr, palette='viridis', s=100)
plt.xlabel('País')
plt.ylabel('Número de Visualizações')
plt.title('Número de Visualizações por País com Correlação com Matrícula no Ensino Superior (%)')
plt.xticks(rotation=75)
plt.colorbar(label='Matrículas no Ensino Superior (%)')
plt.tight_layout()
plt.show()

correlacao = df_corr['video views'].corr(df_corr['Gross tertiary education enrollment (%)'])
print(f"Correlação entre visualizações e Gross tertiary education enrollment (%): {correlacao}")
```

```python
df_categoria_canais = df['channel_type'].value_counts()

df_counts_categoria = pd.DataFrame(df_categoria_canais).reset_index()
df_counts_categoria.columns = ['Channel Type', 'Counts']

print(tabulate(df_counts_categoria, headers='keys', tablefmt='grid', showindex=False))
```

```python
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(df_categoria_canais)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Nuvem de Palavras do Tipo de Canais')
plt.show()
```
